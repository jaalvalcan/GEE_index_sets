```python
import geemap
```

```python
import xee
```

```python
map = geemap.Map(basemap = 'SATELLITE', ee_initialize=False)
map
```

```python
roi = map.draw_last_feature.geometry()
roi
```

```python
biomass =(
    ee.Image("LARSE/GEDI/GEDI04_B_002").select('MU')
)

emb = (
    ee.ImageCollection("GOOGLE/SATELLITE_EMBEDDING/V1/ANNUAL")
    .filterDate('2019','2021')
    .filterBounds(roi)
    .mean()
)

stack = emb.addBands(biomass)
stack
```

```python
import xarray as xr
```

```python
ds = xr.open_dataset(
    stack,
    engine = 'ee',
    crs = 'EPSG:4326',
    scale = 0.01,
    geometry = roi
)
```

```python
ds = ds.squeeze('time').drop_vars('time') * 1

ds
```

```python
df = ds.to_dataframe().dropna()

df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_x = StandardScaler()
scaler_y = StandardScaler()

x = df.drop('MU', axis = 1)
x_scaled = scaler_x.fit_transform(x)
x_scaled

y = df['MU']
y.values.shape

y.values.reshape(-1,1).shape

y_scaled = scaler_y.fit_transform(y.values.reshape(-1,1))
y_scaled

# With only one sample, train_test_split is not appropriate.
# Assign all data to the training set.
x_train, y_train = x_scaled, y_scaled

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x_scaled.shape[1]

model = Sequential()
model.add(Dense(64, activation = 'relu', input_dim = x_scaled.shape[1], ))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse', metrics = ['mae'])

model.fit(
    x_train, y_train,
    epochs = 100,
    batch_size = 32,
    verbose = 1
)
```

```python
df1km = ds.to_dataframe()

scaler_x1km = StandardScaler()
x1km = df1km.drop('MU', axis = 1)
x1km_scaled = scaler_x1km.fit_transform(x1km)
x1km_scaled

df1km['biomass'] =scaler_y.inverse_transform(model.predict(x1km_scaled))

dx = df1km.to_xarray().sortby('lat','lon')

dx['biomass_cor'] = xr.where(dx.biomass < 0, 0.0, dx.biomass)
```
